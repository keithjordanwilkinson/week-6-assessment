```python
def getaroll(participant, personmoney):
    roll = random.choice(wheel)
    print("Player rolled",roll)
    
    if roll == "Lose a turn":
        participant = participant+1
        print("Turn lost")
        updateparticipant(participant, rounds)
        
    elif roll == "Bankrupt":
        personmoney[i] = 0
        participant = participant+1
        print("Turn and Money Lost")
        updateparticipant(participant, rounds)
        
    else: 
        playgame(participant, rounds, hintforguessing, underlineword, personmoney, roll)
    
    return roll


#print(getaroll())
    
```
